## Tratamento de Dados e Múltiplas Bases

### Importando as Bases de Dados

```python
vendas_df = pd.read_csv('Contoso - Vendas - 2017.csv', sep=';')

produtos_df = pd.read_csv('Contoso - Cadastro Produtos.csv', sep=';')

lojas_df = pd.read_csv('Contoso - Lojas.csv', sep=';')

clientes_df = pd.read_csv('Contoso - Clientes.csv', sep=';')

print('VENDAS')
display(vendas_df)

print()

print('PRODUTOS')
display(produtos_df)

print()

print('LOJAS')
display(lojas_df)

print()

print('CLIENTES')
display(clientes_df)
```

### Retirando Colunas Desnecessárias

```python
#CLIENTES

""" #Retirando colunas
clientes_df = clientes_df.drop(['Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10'], axis=1)

display(clientes_df) """

#Pegando colunas
clientes_df = clientes_df[['ID Cliente', 'E-mail']]

print('CLIENTES - TRATADO')
display(clientes_df)

print()

#PRODUTOS
produtos_df = produtos_df[['Nome do Produto', 'ID Produto']]

print('PRODUTOS - TRATADO')
display(produtos_df)

print()

#LOJAS
lojas_df = lojas_df[['ID Loja', 'Nome da Loja']]

print('LOJAS - TRATADO')
display(lojas_df)
```

### Juntando DataFrames

```python
vendas_df = vendas_df.merge(produtos_df, on='ID Produto')
vendas_df = vendas_df.merge(lojas_df, on='ID Loja')
vendas_df = vendas_df.merge(clientes_df, on='ID Cliente')

print('VENDAS - MESCLADO')
display(vendas_df)
```

### Alterando o nome de uma coluna

```python
vendas_df = vendas_df.rename(columns={'E-mail': 'E-mail do Cliente'})

print('VENDAS - RENOMEADO')
display(vendas_df)
```

### 1. Qual cliente comprou mais vezes?

```python
frequencia_clientes = vendas_df['E-mail do Cliente'].value_counts()

print('RESPOSTA EM TEXTO')
display(frequencia_clientes)

print()

#Plotando os primeiros 5 clientes
print('RESPOSTA GRÁFICA')
frequencia_clientes[:6].plot(figsize=(17, 5), yticks=range(0, 100, 5))
```

### 2. Qual loja vendeu mais?

```python
#Agrupando o dataframe
vendas_lojas = vendas_df.groupby('Nome da Loja').sum()

print('TABELA COMPLETA')
display(vendas_lojas)


print()


#Filtrando a exibição
vendas_lojas = vendas_lojas[['Quantidade Vendida']]

print('TABELA FILTRADA')
display(vendas_lojas)


print()


#Ordenando o dataframe
vendas_lojas = vendas_lojas.sort_values('Quantidade Vendida', ascending=False)

print('TABELA ORDENADA EM ORDEM DECRESCENTE')
display(vendas_lojas)


print()


#Plotando um gráfico
vendas_lojas[:6].plot(figsize=(17, 5), kind='bar')


#Pegando o maior valor diretamente
maior_valor = vendas_lojas['Quantidade Vendida'].max()
melhor_loja = vendas_lojas['Quantidade Vendida'].idxmax()

print('MELHOR LOJA: ', melhor_loja)
print('VALOR: ', maior_valor)


print()


print('GRÁFICO') #O gráfico é sempre plotado ao final
```

### 3. Qual loja vendeu menos?

```python
menor_valor = vendas_lojas['Quantidade Vendida'].min()
pior_loja = vendas_lojas['Quantidade Vendida'].idxmin()

print('PIOR LOJA: ', pior_loja)
print('VALOR: ', menor_valor)
```
